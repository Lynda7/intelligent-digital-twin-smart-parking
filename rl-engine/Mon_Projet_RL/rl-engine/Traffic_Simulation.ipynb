```python
import paho.mqtt.client as mqtt
import time
import random

#La Configuration MQTT :
BROKER_ADDRESS = "broker.hivemq.com"
BROKER_PORT = 1883

# Les Topics de communication:
TOPIC_ARRIVAL = "cupcarbon/linda/simulator/entry"
TOPIC_EXIT = "cupcarbon/linda/simulator/exit"
TOPIC_ACTION = "cupcarbon/linda/simulator/action"

# Les variables de Simulation:
parked_cars = []
car_counter = 0

#Les fonction :
def on_connect(client, userdata, flags, rc):
    if rc == 0:
        print("✅ Connecté à MQTT avec succès.")
    else:
        print(f"❌ Échec de la connexion à MQTT. Code de retour: {rc}")

def simulate_10_minute_run_realistic():
    global parked_cars, car_counter

    client = mqtt.Client()
    client.on_connect = on_connect

    try:
        client.connect(BROKER_ADDRESS, BROKER_PORT, 60)
        client.loop_start()

        print("🅿️ Début de la simulation d'un real parking")
        start_time = time.time()
        simulation_duration = 10 * 60  # 10 minutes en secondes

        while (time.time() - start_time) < simulation_duration:
            elapsed_time = time.time() - start_time

            # Définir les probabilités d'arrivée vs départ en fonction du temps
            if elapsed_time < 3 * 60: # Première phase (0-3 min)
                # Forte probabilité d'arrivée
                action = random.choices(['arrive', 'leave'], weights=[0.8, 0.2], k=1)[0]
            elif elapsed_time < 7 * 60: # Deuxième phase (3-7 min)
                # Probabilités équilibrées
                action = random.choices(['arrive', 'leave'], weights=[0.5, 0.5], k=1)[0]
            else: # Troisième phase (7-10 min)
                # Forte probabilité de départ
                action = random.choices(['arrive', 'leave'], weights=[0.2, 0.8], k=1)[0]

            if action == 'arrive' or not parked_cars:
                car_counter += 1
                car_id = f"Voiture_{car_counter}"
                parked_cars.append(car_id)
                client.publish(TOPIC_ARRIVAL, car_id)
                print(f"🚗 Événement d'arrivée publié: {car_id}")
            elif action == 'leave' and parked_cars:
                car_to_leave = random.choice(parked_cars)
                parked_cars.remove(car_to_leave)
                client.publish(TOPIC_EXIT, car_to_leave)
                print(f"👋 Événement de départ publié: {car_to_leave}")

            # Attendre 1 seconde avant la prochaine action
            time.sleep(1)

        print("\n🏁 Simulation terminée. Publication du message de fin.")
        client.publish(TOPIC_ACTION, "STOP_SIMULATION")

    except KeyboardInterrupt:
        print("\nℹ️ Simulation arrêtée manuellement.")
    except Exception as e:
        print(f"Une erreur s'est produite: {e}")
    finally:
        client.loop_stop()
        client.disconnect()
        print("Déconnecté.")

if __name__ == "__main__":
    simulate_10_minute_run_realistic()
```
